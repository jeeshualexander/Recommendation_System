# Extracting Zomato Reviews

```python
#!pip install selenium
import pandas as pd
import bs4
import selenium
import requests
import re
```

```python
def get_info(review, selector, attribute):
    try:
        tag = review.find_element_by_css_selector(selector)
        tag = tag.get_attribute(attribute)
    except:
        tag = 0
    return tag
```

```python
from selenium import webdriver
browser = webdriver.Chrome('chromedriver.exe')
browser.get("https://www.zomato.com/bangalore/great-food-no-bull")
```

```python
finest = browser.find_elements_by_css_selector('div.relative.top-res-box.entity-ads-snippet-track')
print(len(finest))
```

```python
df_restname = pd.DataFrame(columns = ['RestaurantName', 'Url'])
for restn in finest:
    restaurant = get_info(restn, "div.res_title.zblack.bold.nowrap", "title")
    url = get_info(restn, "a.jumbo_track_collections", "href")
    curr_rest_name = {
        'RestaurantName': restaurant,
        'Url': url
    }
    df_restname = df_restname.append(curr_rest_name, ignore_index = True)
print(df_restname.head())
print(df_restname.shape)
a = df_restname[df_restname['RestaurantName']==0]
for url in a['Url']:
    browser.get(url)
    outlets = browser.find_elements_by_css_selector('div.card.search-snippet-card.search-card')
    for restn in outlets:
        restaurant = get_info(restn, "a.result-title.hover_feedback.zred.bold.ln24.fontsize0", "title")
        url = get_info(restn, "a.result-title.hover_feedback.zred.bold.ln24.fontsize0", "href")
        curr_rest_name = {
                'RestaurantName': restaurant,
                'Url': url
            }
        df_restname = df_restname.append(curr_rest_name, ignore_index = True)
df_restname = df_restname[df_restname.RestaurantName!=0]
print(df_restname.shape)
```

```python
df_rest_rev = pd.DataFrame(columns = ['UserId', 'Username', 'Restaurant', 'Address', 'Rating', 'Comment'])
for url in df_restname['Url']:
    browser.get(url)
    while(True):
        try:
            more_sel = 'div.ui.segment.clearfix.zs-load-more.res-page-load-more'
            load_more_button = browser.find_element_by_css_selector(more_sel)
            load_more_button.click()
        except:
            break
    zom_rev = browser.find_elements_by_css_selector('div.ui.segment.clearfix.brtop')
    rest_name = browser.find_elements_by_css_selector('div.ui.segment.res-header-overlay.vr')
    print(len(zom_rev))
    print(len(rest_name))
    print('-----------------------------------------------------')
    for review in zom_rev:
        userid = get_info(review, "a[data-entity_id]", "data-entity_id")
        username = get_info(review, "a[data-entity_id]", "innerHTML")
        username = username.replace("\n","").strip()
        rating = get_info(review, "div[aria-label]", "aria-label")
        rating = rating.replace("Rated","").strip()
        comment = get_info(review, "div.rev-text.mbot0", "innerHTML")
        comment = re.sub('<.*?>', '', comment)
        comment = comment.replace("Rated&nbsp;","").strip()
        for restname in rest_name:
            restaurant = get_info(restname, "a.ui.large.header.left", "title")
            address = get_info(restname, "a.left.grey-text.fontsize3", "innerHTML")
            address = address.replace("\n","").strip()
        curr_review = {
            'UserId': userid,
            'Username': username,
            'Restaurant': restaurant,
            'Address': address,
            'Rating': rating,
            'Comment': comment
        }
        df_rest_rev = df_rest_rev.append(curr_review, ignore_index = True)
print(df_rest_rev.shape)
df_rest_rev = df_rest_rev.drop_duplicates().reset_index(drop=True)
print(df_rest_rev.shape)
```

# Recomendation System

```python
zomato_reviews = pd.read_csv("C:/Users/jeeshu/Desktop/Datasets/Term2/Machine_Learning/Reviews.csv")
```

```python
zomato_reviews.head()
```

```python
user_matrix = zomato_reviews.pivot_table(index='UserId', columns='Restaurant', values='Rating')
```

```python
from sklearn.metrics.pairwise import cosine_similarity    #To find cosine simalarity between users
```

```python
ui_matrix = user_matrix.apply(lambda v: v - v.mean(), axis=1).fillna(0)
```

```python
sim_matrix = pd.DataFrame(cosine_similarity(ui_matrix), index=ui_matrix.index, columns=ui_matrix.index)   #Gets the cosine value
```

```python
def rest(userid):
    pred = pd.DataFrame(columns=['Restaurant', 'predicted_rating'])
    neigh_user = sim_matrix[userid].drop(userid).sort_values(ascending = False).head(7).index
    for Restaurant, rating in user_matrix.loc[userid].items():     
        if pd.isnull(rating):
            mid_rating = user_matrix.loc[neigh_user, Restaurant]
            avg_mid_rating = mid_rating.mean()
            pred = pred.append({'Restaurant':Restaurant, 
                                'predicted_rating':avg_mid_rating}, ignore_index=True)
    
    restaurant = pred.sort_values('predicted_rating',ascending=False).head(3)
    return restaurant
```
